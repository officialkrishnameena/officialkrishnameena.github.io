```python
from sklearn.model_selection import train_test_split#Splitting the data in training, vaalidation and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=42)
```

```python
class MyKNeighborsClassifier():

    def __init__(self, n_neighbors=1):
        self.n_neighbors=n_neighbors

    def fit(self, X, y):

        n_samples = X.shape[0]
        # number of neighbors can't be larger then number of samples
        if self.n_neighbors > n_samples:
            raise ValueError("Number of neighbors can't be larger then number of samples in training set.")

        # X and y need to have the same number of samples
        if X.shape[0] != y.shape[0]:
            raise ValueError("Number of samples in X and y need to be equal.")

        # finding and saving all possible class labels
        self.classes_ = np.unique(y)

        self.X = X
        self.y = y

    def predict(self, X_val):

        # number of predictions to make and number of features inside single sample
        n_predictions, n_features = X_val.shape

        # allocationg space for array of predictions
        predictions = np.empty(n_predictions, dtype=int)

        # loop over all observations
        for i in range(n_predictions):
            # calculation of single prediction
            predictions[i] = single_prediction(self.X, self.y, X_val[i, :], self.n_neighbors)

        return(predictions)
```

```python
def single_prediction(X, y, x_train, k):

    # number of samples inside training set
    n_samples = X.shape[0]

    # create array for distances and targets
    distances = np.empty(n_samples, dtype=np.float64)

    # distance calculation
    for i in range(n_samples):
        distances[i] = (x_train - X[i]).dot(x_train - X[i])

    # combining arrays as columns
    distances = sp.c_[distances, y]
    # sorting array by value of first column
    sorted_distances = distances[distances[:,0].argsort()]
    # selecting labels associated with k smallest distances
    targets = sorted_distances[0:k,1]

    unique, counts = np.unique(targets, return_counts=True)
    return(unique[np.argmax(counts)])
```

```python
# Instantiate learning model
my_classifier = MyKNeighborsClassifier(n_neighbors=3)

# Fitting the model
my_classifier.fit(X_val, y_val)

# Predicting the Test set results
my_y_pred = my_classifier.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_val, my_y_pred)*100
print('Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.utils import resample
from sklearn.metrics import accuracy_score

n_folds = 5

# Initializing a list to store accuracy scores for each fold
fold_accuracies = []

# Creating a StratifiedKFold object for stratified sampling
stratified_kf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)

# Loop through the folds
for train_index, test_index in stratified_kf.split(X, y):
    # Split data into training and test sets for the current fold
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    bootstrap_accuracies = []

    # Number of bootstrap samples
    bootstrap_samples = 5
    # Perform bootstrapping
    for _ in range(bootstrap_samples):
        # Resample the training data with replacement
        X_train_bootstrap, y_train_bootstrap = resample(X_train, y_train, random_state=42)

        # Scale the features
        X_train_bootstrap = scaler.fit_transform(X_train_bootstrap)
        X_test_scaled = scaler.transform(X_test)

        # Training the model with KNN
        model = KNeighborsClassifier(n_neighbors=5)
        model.fit(X_train_bootstrap, y_train_bootstrap)

        # Evaluating the model on the test set
        y_pred = model.predict(X_test_scaled)

        # Calculating accuracy and storing it
        accuracy = accuracy_score(y_test, y_pred)
        bootstrap_accuracies.append(accuracy)

    # Calculatig the mean accuracy
    fold_accuracy = np.mean(bootstrap_accuracies)
    fold_accuracies.append(fold_accuracy)

# Calculating the overall mean accuracy across all folds
mean_accuracy = np.mean(fold_accuracies)*100

# Printing the mean accuracy
print("Mean Accuracy:",mean_accuracy,'%')
```
